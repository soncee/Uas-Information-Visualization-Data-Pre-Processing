```python
bins = [18, 25, 40, 50, 60, 70, 120]
labels = ['18-24', '25-39', '40-49', '50-59', '60-69', '70+']
df4['agerange'] = pd.cut(df4.age, bins, labels = labels,include_lowest = True)
df4
```
